# TIC TAC TOE GAME
### HERE ARE THE REQUIREMENTS:
###### ** 2 players should be able to play the  game (both sitting at the same  computer)
###### ** The board should be printed out every time a players makes  a move
###### ** You should be able to accept input of the player position and then place a symbol '-' on the board

```python
## board
board = ['-', '-', '-', '-', '-', '-', '-', '-', '-']
print("WELCOME TO TIC TAC TOE ")
# Display the board
def display_board():
    print('     |     |')
    print('  {}  |  {}  |  {}'.format(board[0], board[1], board[2]))
    print('_____|_____|_____')
    print('     |     |')
    print('  {}  |  {}  |  {}'.format(board[3], board[4], board[5]))
    print('_____|_____|_____')
    print('     |     |')
    print('  {}  |  {}  |  {}'.format(board[6], board[7], board[8]))
    print('     |     |')


# Check the possibility of the winner
def check_win(board, mark):
    # Check for horizontal win
    for i in range(0, 9, 3):
        if board[i] == board[i+1] == board[i+2] == mark:
            return True
    # Check for vertical win
    for i in range(0, 3):
        if board[i] == board[i+3] == board[i+6] == mark:
            return True
    # Check for diagonal win
    if board[0] == board[4] == board[8] == mark:
        return True
    elif board[2] == board[4] == board[6] == mark:
        return True
    return False

# Get the valid position from the user
def get_valid_position(board, mark):
    while True:
        try:
            pos = int(input(f'{mark}: Enter the position (1-9): '))
            if pos < 1 or pos > 9:
                print('Invalid position. Enter a number between 1-9.')
            elif board[pos-1] != '-':
                print('Position already occupied. Choose another position.')
            else:
                return pos
        except ValueError:
            print('Invalid input. Enter a number between 1-9.')

# Get the player names
player1_name = input('Enter player 1 name: ')
player2_name = input('Enter player 2 name: ')

# Set player marks
player1_mark = 'X'
player2_mark = 'O'

# Start the game
display_board()
for i in range(9):
    if i % 2 == 0:
        pos = get_valid_position(board, player1_name)
        board[pos-1] = player1_mark
        display_board()
        if check_win(board, player1_mark):
            print(f'Congratulations {player1_name}, you won!')
            break
    else:
        pos = get_valid_position(board, player2_name)
        board[pos-1] = player2_mark
        display_board()
        if check_win(board, player2_mark):
            print(f'Congratulations {player2_name}, you won!')
            break
    if i == 8:
        print('Game over. It\'s a draw.')
```
